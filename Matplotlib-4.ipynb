### Lines and Span Rectangles

```python
#to draw a line between (x1, x2) to (y1, y2) just specify points:
plt.plot([5,5], [10,0]) #vertical blue line
plt.plot([0,10], [5,5]) #horizontal line
plt.show()
```

this approach would be tedious
for drawing a large number of disconnected lines so for horizontal and vertical lines
there are a pair of convenient methods, ax.hlines and ax.vlines. ax.hlines takes
mandatory arguments, y, xmin, xmax, and draws horizontal lines with y-coordinates at
each of the values given by the sequence y (if y is passed as a scalar, a single line is
drawn).

```python
plt.hlines(5, 1, 10) #for horizontal lines: plt.hlines(y_ordinate, xmin, xmax)
```

```python
plt.vlines(5, 1, 10) #similar for vertical lines
```

```python
fig, ax = plt.subplots()
for i in range(10):
    ax.vlines(i, i-5, i+5)
    ax.hlines(i, i-5, i+5, 'r') #this was a random pattern but did good
```

```python
#drawing circle with those lines:
#requirement for line height first 0, then increase to max value and then in same pattern decrease

fig, ax = plt.subplots(figsize = (5,5))
x = np.linspace(-1, 1, 100)
y = np.sqrt(1-x**2)
ax.hlines(y, -x, x)
ax.hlines(-y, -x, x)

#lets draw box around circle
ax.vlines([-1,1], -1, 1,'k', '-.') #first list of x pos, then we can use scalar -1 representing ymin and y max for two lines
#two vertical lines at abscissa -1 and 1 is represented by that list

ax.set_xticks([])#remove both ticks and labels
ax.set_yticklabels([]) #removes tick labels only
```

On static plots such as figures for printing, ax.hlines and ax.vlines work well, but
note that the line limits don’t change upon changing the axes’ limits in an interactive
plot. 

There are two further methods, `ax.axhline` and `ax.axvline`, which simply plot
a horizontal or vertical line across the axis, whatever its current limits. ax.axhline
takes arguments `y, xmin, xmax`, but these must be scalar values (so multiple lines require
repeated calls) and xmin, xmax are given in fractional coordinates such that 0 is the far
left of the plot and 1 the far right. 

```python
fig, ax = plt.subplots()
ax.axhline(100) #axis horizontal line spannning whole image 
ax.axvline (5, 0.4, 0.6, c='b', lw=4, ls='--')
plt.show()
#this axhline has limit of 0 to 1 only
```

`axhspan` and `axvspan` is used to draw rectanges spanning specified ranges

```python
fig, ax = plt.subplots()
ax.axhspan(5,8,xmin=.2, xmax=.8) 
ax.set_ylim(-10,10)
plt.show()
```

Here are list of arguments that can be used on axhspan and axvspan and other patch like objects



```python
#now lets draw a nice strip of weave like structures:
i=0; j=0
fig, ax = plt.subplots(figsize=(5,5))
for i in range(10):
    ax.axhspan(i,i+.5, fc='y',ec='g')
    for j in range(i, 10):
        ax.axvspan(j,j+.5, fc='g', ec='k')

ax.set_xticks([])
ax.set_yticks([])
plt.show()
    
```

A representation of the electromagnetic spectrum, 250–1000 nm


```python
plt.figure()
```

```python
lmin , lmax = 250, 1000
x = np.linspace(lmin , lmax , 1000)
# A wave with a smoothly increasing wavelength .
wv = (np.sin (10 * np.pi * x / (lmax+lmin -x)))[::-1]

fig, ax = plt.subplots()
ax.plot(x, wv , c='k', lw =2)
ax.set_xlim (250 , 1000)
ax.set_ylim (-2, 2)

# Label and delimit the different regions of the electromagnetic spectrum .
ax.text(310 , 1.5, 'UV', color='k', fontdict ={'fontsize': 20})
ax.text (530 , 1.5, 'Visible', color='k', fontdict ={'fontsize': 20})
ax.annotate ('', (400 , 1.3) , (750 , 1.3) , arrowprops ={'arrowstyle': '<|-|>','color': 'k', 'lw': 2})

ax.text(860 , 1.5, 'IR', color='k', fontdict ={'fontsize': 20})

ax.axvline (400 , -2, 2, c='k', ls='--')
ax.axvline (750 , -2, 2, c='k', ls='--')
# Horizontal "axis" across the center of the wave.
ax.axhline(c='k')
# Remove the y-axis ticks and labels; label the x-axis.
ax.yaxis. set_visible (False)
ax. set_xlabel (r'$\lambda \;/\mathrm{nm}$')

# Finally , add some colorful rectangles representing a rainbow in the
# visible region of the spectrum.
# Dictionary mapping of wavelength regions (nm) to approximate RGB values.
rainbow_rgb = { (400 , 440): '#8b00ff', (440 , 460): '#4b0082',(460 , 500): '#0000ff', (500 , 570): '#00ff00',(570 , 590): '#ffff00', (590 , 620): '#ff7f00',(620 , 750): '#ff0000'}
for wv_range , rgb in rainbow_rgb .items ():
    ax.axvspan (* wv_range , color=rgb , ec='none', alpha =.75)
plt.show ()
```
