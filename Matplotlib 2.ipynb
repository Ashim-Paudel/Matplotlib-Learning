````python
fig = plt.figure("Contrast between x cosx and its fourier series", figsize = (10,7), facecolor="white", edgecolor="white")
ax = fig.add_subplot()
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
func = ax.plot(x, x*np.cos(x), label = "x cosx")
fourier = -np.sin(x)/2 
for n in range(2,50):
    fourier += 2*(-1)**(n+1)*np.sin(x)/(n**2-1)
approx = ax.plot(x, fourier, label = "Fourier")
plt.title("Contrast between x cosx and its fourier series")
plt.legend(loc="upper center")
plt.show()
````

### comparison between fourier series of:
$|\sin x|$  and $| \cos x|$ in range $-\pi \le x \le \pi$



````python
fig = plt.figure(figsize=(12,6), facecolor="white", edgecolor="white")
x = np.linspace(-np.pi, np.pi, 1000)
ax = fig.add_subplot()
abs_sin = ax.plot(x, abs(np.sin(x)), label="|sinx|")

fourier = 2/np.pi
for n in range(1,1000):
    fourier += -2/np.pi * ((-1)**n+1)/(n**2+1) * np.cos(n*x)

fourier_graph = ax.plot(x, fourier, label="fourier")
plt.legend(loc= 'best')
plt.title("Comparison of |sinx| and its fourier approximation in range -pi to pi")
plt.savefig("images/mod_sinx_and_fourier.png")
plt.show()
````

````python
#now for mod cosx

fig = plt.figure(figsize=(12,6), facecolor="white", edgecolor="white")
x = np.linspace(-np.pi, np.pi, 1000)

ax = fig.add_subplot()
abs_sin = ax.plot(x, abs(np.cos(x)), label="|cos|")

fourier = 2/np.pi
for n in range(2,15):
    fourier += 2/np.pi * (np.sin((n+1)*np.pi/2)/(n+1)+np.sin((n-1)*np.pi/2)/(n-1)) * np.cos(n*x)

fourier_graph = ax.plot(x, fourier, label="fourier")
plt.legend(loc= 'best')
plt.title("Comparison of |cosx| and its fourier approximation in range -5pi to 5pi")
plt.savefig("images/mod_cosx_and_fourier.png")
plt.show()
````


### Scatter plots
To explore the correlation between birth rate, life expectancy and per
capita income, we may use a scatter plot. The marker sizes are set in proportion to the
countries per-capita gross domestic product (GDP) but have to be scaled a little so they
dont get too large

````python
# some extracted data

countries = ['Brazil', 'Madagascar', 'S. Korea', 'United States','Ethiopia', 'Pakistan', 'China', 'Belize']
# Birth rate per 1000 population .
birth_rate = np.array([16.4 , 33.5 , 9.5, 14.2 , 38.6 , 30.2 , 13.5 , 23.0])
# Life expectancy at birth , years.
life_expectancy = np.array([73.7 , 64.3 , 81.3 , 78.8 , 63.0 , 66.4 , 75.2 , 73.7])
# Per person income fixed to US Dollars in 2000.
GDP = np.array ([4800 , 240, 16700 , 37700 , 230, 670, 2640 , 3490])


fig, ax = plt.subplots()
#we can instead have initialized two objects fig and ax seperately if we needed to pass arguments

ax.scatter(birth_rate, life_expectancy, c = range(len(countries)), s=GDP/25)
ax.set_xlim(5, 45)
ax.set_ylim(60,85)
ax.set_xlabel("Birth rate")
ax.set_ylabel("Life expectancy")
ax.grid(True, c='.6') #to set grid to true and c to assing colors to grid lines

plt.title("Correlation between birthrate, life expectancy and gdp", fontsize=15, fontname="times new roman" )
plt.show()
````

### Another practice
The historical populations of five US cities are given in the files boston
.tsv, houston.tsv, detroit.tsv, san_jose.tsv, phoenix.tsv as tab-separated columns
of (year, population). They can be downloaded from https://scipython.com/eg/baf.
The following program plots these data on one set of axes with a different line style
for each.

````python
cities = ['Boston', 'Houston', 'Detroit', 'San Jose', 'Phoenix']
linestyle = [{'ls':'-'}, {'ls':':'}, {'ls':'--'}, {'ls':'-.'}, {'dashes':[2,3,2,4,2,3]}]

fig,ax = plt.subplots()
for i,city in enumerate(cities):
    filename = f"datas/{city.lower().replace(' ', '_')}.tsv"
    yr , pop = np.loadtxt(filename , unpack=True)
    line = ax.plot(yr ,pop/1e6, label=city, **linestyle[i])

ax.set_xlabel("Year")
ax.set_ylabel("Population (millions)")
ax.set_title("The historical populations of five US cities")
ax.legend(loc="upper left")
plt.show()
````

### Using fonts

the syntax is:

`ax.title("tile_xyz", fontsize = 18, fontname = "times new roman", color="blue")`

To use the same font properties for all text elements, it is easiest to set Matplotlibs rc settings using a dictionary of values. This involves a separate import first:

```
from matplotlib import rc
font_properties = {'family' : 'monospace',
'weight' : 'bold',
'size' : 22}
rc('font', ** font_properties )
# All text will now be rendered in 22-point , bold monospace in plots.
```

*rc means runtime configuration*


reference: https://matplotlib.org/stable/tutorials/introductory/customizing.html

````python
import numpy as np
import matplotlib .pyplot as plt
# Initial value of y at t=0, lifetime (s).
N, tau = 10000 , 28
# Maximum time to consider (s).
tmax = 100
# A suitable grid of time points , and the exponential decay itself.
t = np.linspace (0, tmax , 1000)
y = N * np.exp(-t/tau)
fig = plt.figure ()
ax = fig. add_subplot ()
ax.plot(t, y)
# The number of lifetimes that fall within the plotted time interval .
ntau = tmax // tau + 1
# xticks at 0, tau , 2*tau , ..., ntau*tau; yticks at the corresponding y-values.
xticks = [i*tau for i in range(ntau )]
yticks = [N * np.exp(-i) for i in range(ntau )]
ax. set_xticks (xticks)
ax. set_yticks (yticks)
# xtick labels: 0, tau , 2tau , ...
xtick_labels = [r'$0$', r'$\tau$'] + [r'${}\ tau$'.format(k) for k in range(2, ntau )]
ax. set_xticklabels ( xtick_labels )
# Corresponding ytick labels: N, N/e, N/2e, ...
ytick_labels = [r'$N$', r'$N/e$'] + [r'$N /{} e$'.format(k) for k in range(2, ntau )]
ax. set_yticklabels ( ytick_labels )
ax. set_xlabel (r'$t \;/\ mathrm{s}$')
ax. set_ylabel (r'$y$')
ax.grid ()
plt.show ()
````
