### comparison between fourier series of:
$|\sin x|$  and $| \cos x|$ in range $-\pi \le x \le \pi$







````python
fig = plt.figure(figsize=(12,6), facecolor="white", edgecolor="white")
x = np.linspace(-np.pi, np.pi, 1000)
ax = fig.add_subplot()
abs_sin = ax.plot(x, abs(np.sin(x)), label="|sinx|")

fourier = 2/np.pi
for n in range(1,1000):
    fourier += -2/np.pi * ((-1)**n+1)/(n**2+1) * np.cos(n*x)

fourier_graph = ax.plot(x, fourier, label="fourier")
plt.legend(loc= 'best')
plt.title(r"Comparison of $|\sin x|$ and its fourier approximation in range $-\pi$ to $\pi$")
plt.savefig("images/mod_sinx_and_fourier.png")
plt.show()
````

````python
#now for mod cosx

fig = plt.figure(figsize=(12,6), facecolor="white", edgecolor="white")
x = np.linspace(-np.pi, np.pi, 1000)

ax = fig.add_subplot()
abs_sin = ax.plot(x, abs(np.cos(x)), label=r"$|\cos x|$")

fourier = 2/np.pi
for n in range(2,15):
    fourier += 2/np.pi * (np.sin((n+1)*np.pi/2)/(n+1)+np.sin((n-1)*np.pi/2)/(n-1)) * np.cos(n*x)

fourier_graph = ax.plot(x, fourier, label="fourier")
plt.legend(loc= 'best')
plt.title(f"Comparison of $|cosx|$ and its fourier approximation in range $-5\pi$ to $5\pi$")
plt.savefig("images/mod_cosx_and_fourier.png")
plt.show()
````


### Scatter plots
To explore the correlation between birth rate, life expectancy and per
capita income, we may use a scatter plot. The marker sizes are set in proportion to the
countries’ per-capita gross domestic product (GDP) but have to be scaled a little so they
don’t get too large

````python
# some extracted data

countries = ['Brazil', 'Madagascar', 'S. Korea', 'United States','Ethiopia', 'Pakistan', 'China', 'Belize']
# Birth rate per 1000 population .
birth_rate = np.array([16.4 , 33.5 , 9.5, 14.2 , 38.6 , 30.2 , 13.5 , 23.0])
# Life expectancy at birth , years.
life_expectancy = np.array([73.7 , 64.3 , 81.3 , 78.8 , 63.0 , 66.4 , 75.2 , 73.7])
# Per person income fixed to US Dollars in 2000.
GDP = np.array ([4800 , 240, 16700 , 37700 , 230, 670, 2640 , 3490])


fig, ax = plt.subplots()
#we can instead have initialized two objects fig and ax seperately if we needed to pass arguments

ax.scatter(birth_rate, life_expectancy, c = range(len(countries)), s=GDP/25)
ax.set_xlim(5, 45)
ax.set_ylim(60,85)
ax.set_xlabel("Birth rate")
ax.set_ylabel("Life expectancy")
ax.grid(True, c='.6') #to set grid to true and c to assing colors to grid lines

plt.title("Correlation between birthrate, life expectancy and gdp", fontsize=15, fontname="times new roman" )
plt.show()
````

### Another practice
The historical populations of five US cities are given in the files boston
.tsv, houston.tsv, detroit.tsv, san_jose.tsv, phoenix.tsv as tab-separated columns
of (year, population). They can be downloaded from https://scipython.com/eg/baf.
The following program plots these data on one set of axes with a different line style
for each.

````python
cities = ['Boston', 'Houston', 'Detroit', 'San Jose', 'Phoenix']
linestyle = [{'ls':'-'}, {'ls':':'}, {'ls':'--'}, {'ls':'-.'}, {'dashes':[2,3,2,4,2,3]}]

fig,ax = plt.subplots()
for i,city in enumerate(cities):
    filename = f"datas/{city.lower().replace(' ', '_')}.tsv"
    yr , pop = np.loadtxt(filename , unpack=True)
    line = ax.plot(yr ,pop/1e6, label=city, **linestyle[i])

ax.set_xlabel("Year")
ax.set_ylabel("Population (millions)")
ax.set_title("The historical populations of five US cities")
ax.legend(loc="upper left")
plt.show()
````

**Choosing the right font family**
- Choose the font family (‘serif’ or ‘sans-serif’ to match the paper - most of the time, ‘serif’ is best):
- `plt.rc('font', family='serif')`

````python
import numpy as np
import matplotlib .pyplot as plt
# Initial value of y at t=0, lifetime (s).
N, tau = 10000 , 28
# Maximum time to consider (s).
tmax = 100
# A suitable grid of time points , and the exponential decay itself.
t = np.linspace (0, tmax , 1000)
y = N * np.exp(-t/tau)
fig = plt.figure ()
ax = fig. add_subplot ()
ax.plot(t, y)
# The number of lifetimes that fall within the plotted time interval .
ntau = tmax // tau + 1
# xticks at 0, tau , 2*tau , ..., ntau*tau; yticks at the corresponding y-values.
xticks = [i*tau for i in range(ntau )]
yticks = [N * np.exp(-i) for i in range(ntau )]
ax. set_xticks (xticks)
ax. set_yticks (yticks)
# xtick labels: 0, tau , 2tau , ...
xtick_labels = [r'$0$', r'$\tau$'] + [r'${}\tau$'.format(k) for k in range(2, ntau )]
ax. set_xticklabels ( xtick_labels )
# Corresponding ytick labels: N, N/e, N/2e, ...
ytick_labels = [r'$N$', r'$N/e$'] + [r'$N /{} e$'.format(k) for k in range(2, ntau )]
ax. set_yticklabels ( ytick_labels )
ax. set_xlabel (r'$t \;/\ mathrm{s}$')
ax. set_ylabel (r'$y$')
ax.grid ()
plt.show ()
````

## Multiple Subplots

### Basis method

Multiple subplots can be created by creating different axes object of fig object by specifying number of rows, columns of total subplots and the position of that axis in subplot.

*eg: for creating 2 subplots*

```
fig = plt.figure()
ax_1 = fig.add_subplot(121) #121 represents: (1*2)subplots and last 1 gives position for ax_1
ax_2 = fig.add_subplot(122)
```

**We are using two new things too**:
- `fig.suptitle()` This allows us to add title to whole plot, instead of using `plt.title()` which makes rather unmanaged title.

- `fig.tight_layout()` is used to make good fitting and appropriate spacing betweeen those subplots. Moreover, the parameters that we can use in tight_layout are padding parameters which are as: `fig.tight_layout(pad, w_pad, h_pad)`

````python
#now let's create 3*3 subplots using python loop and draw graph of x^n

fig = plt.figure("comparison of x exponents of natural number", figsize=(10,7)) #figsize allots total size of area where plots can be made
axes = list(range(9)) #to store each axes object
lines = list(range(9))
for n in range(9):
    axes[n] = fig.add_subplot(int(f"33{n+1}")) #specifying location of each axis object
    line = axes[n].plot(x, x**n, label=fr"$x^{n}$")
    # axes[n].set_title(fr"$x^{n}$", fontsize=14)
    axes[n].legend(loc="upper center")
    
fig.tight_layout(pad=4, w_pad=1, h_pad=1) #to ensure good spacing of all figures
fig.suptitle(r"Comparison of graph of $x^n$ for $n=0$ to $n=8$.", fontsize = 15)
plt.show()
    
````

### Other ways of defining multiple subplots

- Alternatively, to create a figure and add all its subplots to it at the same time, call `plt.subplots`, which takes arguments `nrows` and `ncols`in addition to previously used arguments like `figsize`, `facecolor`,`edgecolor` etc. and returns a Figure and an array of Axes objects, which can be indexed for each individual axis:


````python
fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(10,7), tight_layout = True) #we are using now all things at a place instead of loop

# in order to access each subplot, use martix method
n = 0
for i in range(3):
    for j in range(3):
        ax[i][j].plot(x, x**(n), label=fr"$x^{n}$", c='bgrcmykbg'[n])
        ax[i][j].legend(loc = "upper center")
        n+=1
fig.suptitle(r"Comparison of graph of $x^n$ for $n=0$ to $n=8$.", fontsize = 15)
plt.show()
````

````python
# Cross - sectional area of bar in m^3 , heat added at x = 0 in J.
A, H = 1.e-4, 1.e3
# Temperature in K at t = 0.
theta0 = 300
# Metal element symbol , specific heat capacities per unit volume (J.m -3.K-1), Thermal diffusivities (m^2/s ) for Cu and Fe.
metals = np.array ([('Cu', 3.45e7 , 1.11e-4), ('Fe', 3.50e7 , 2.3e-5)] ,dtype =[('symbol', '|S2'), ('cp', 'f8'), ('D', 'f8')])
#formula is : theta = H*np.exp(-x**2/(4*D*t))/(Cp*A*np.sqrt(D*t*4*np.pi))
fig,ax = plt.subplots(nrows=3, ncols=2, tight_layout = True, figsize=(8,8), facecolor='white', edgecolor='white', dpi=100)
fig.suptitle("Numerical solutions of the one-dimensional diffusion equation\n for the temperatures of two metal bars.", fontsize=16)
time = (1e-2, 0.1, 1)
x = np.linspace(-0.05, 0.05, 1000)

for i in range(3):
    t = time[i]
    for j in range(2):
        D = metals['D'][j]
        Cp = metals['cp'][j]
        theta = theta0+ H*np.exp(-x**2/(4*D*t))/(Cp*A*np.sqrt(D*t*4*np.pi))
        
        ax[i][j].plot(x*100,theta, label=metals['symbol'][j].decode('utf8'))
        ax[i][j].set_title(fr"${metals['symbol'][j].decode('utf8')}, T = {t}s$")
        #ax[i][j].set_xlim(-4,4)
        ax[i][j].set_xlabel(fr"$x / cm$", fontsize=12)
        ax[i][j].set_ylabel(fr"$\theta /K$", fontsize=12)
        ax[i][j].legend()
    n += 10

plt.savefig('images/one-dimensional-diffusion-equation.png', dpi=100)
plt.show()
````

Now, adjusting sine wave stacking each other from top to bottom\

This code generates a figure of 10 subplots depicting the graph of
sin(nπx) for n = 0, 1, . . . , 9. The subplot spacing is configured so that they “run into”
each other vertically:

````python
fig, ax = plt.subplots(nrows=10, ncols=1, figsize=(7,10))
fig.subplots_adjust(hspace=0)

x=np.linspace(0,1,1000)
for i in range(10):
    ax[i].plot(x, np.sin((10-i)*x*np.pi), 'k', lw=2)
plt.show()
````
